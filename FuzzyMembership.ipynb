```python
import skfuzzy as fuzzy
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
triangle_fz = fuzzy.trimf(x, [1, 5, 10])
triangle_fz
```

```python
plt.plot(x, triangle_fz) 
plt.xlabel('x') 
plt.ylabel('degree of membership')
plt.title('membership function of triangle')
```

```python
trapezoid_fz = fuzzy.trapmf(x, [1, 4, 7, 10])
trapezoid_fz
```

```python
plt.plot(x, trapezoid_fz) 
plt.xlabel('x') 
plt.ylabel('degree of membership')
plt.title('membership function of trapezoid')
```

```python
x = np.arange(1,11, 0.1)
x
```

```python
gaussian_fz = fuzzy.gaussmf(x, 5, 1)
gaussian_fz
```

```python
plt.plot(x, gaussian_fz) 
plt.xlabel('x') 
plt.ylabel('degree of membership')
plt.title('membership function of gaussian with mean = 5 and sigma = 1')
```

```python
gaussian2_fz = fuzzy.gaussmf(x, 5, 0.4)
gaussian2_fz
```

```python
plt.plot(x, gaussian2_fz) 
plt.xlabel('x') 
plt.ylabel('degree of membership')
plt.title('membership function of gaussian with mean = 5 and sigma = 0.4')
```

```python
bellshaped_fz = fuzzy.gbellmf(x, 1, 2, 5)
bellshaped_fz
```

```python
plt.plot(x, bellshaped_fz) 
plt.xlabel('x') 
plt.ylabel('degree of membership')
plt.title('membership function of generalised bellshaped with centre = 5, width = 1 and slope = 2.5')
```
